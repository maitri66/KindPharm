# Importing Libraries

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

# Search for a target protein

## Target search for Coronavirus

```python
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

## Select and retrieve bioactivity data for SARS coronavirus 3C-like proteinase (7th entry)

```python
## We consider the target protein as 'SINGLE PROTEIN' and assign the seventh entry to the selected_target variable.

selected_target = targets.target_chembl_id[6]
selected_target
```

```python
## Retrieve biactivity data for coronavirus 3C-like proteinase (CHEMBL3927) that are reported as IC50 values in nM unit.

activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

IC50 stands for "half maximal inhibitory concentration," and it is a measure used in pharmacology to describe the potency of a substance in inhibiting a specific biological or biochemical function. The IC50 value represents the concentration of a drug or chemical at which 50% inhibition of the biological response is observed.

In the context of drug development and pharmacology, the IC50 is commonly used to assess the effectiveness of a drug in inhibiting the activity of a particular enzyme, receptor, or other molecular target. It's a quantitative measure that helps researchers and pharmacologists compare the inhibitory effects of different compounds.

For example, if a drug has a low IC50, it means that a relatively small amount of the drug is needed to achieve a significant inhibitory effect. On the other hand, a higher IC50 indicates that a higher concentration of the drug is required for the same level of inhibition.

# Handling missing data

```python
## We have to check if any compunds have the missing value for the 'standard_value' column and then we have to drop it.

df2 = df[df.standard_value.notna()]
df2

## Apparantly for this dataset, there is no missing data, use the code for data where there is missing values of 'standard_value'
```

# Data pre-processing of the bioactivity data

## Labelling compunds as either being active, inactive or intermediate

```python
## The bioactivity data is in the IC50 unit.
## > Compounds with values less than 1000 nM are 'Active'.
## > Compounds with values more than 10,000 nM are 'Inactive'.
## > Compounds with values between 1000 nM and 10,000 nM are 'Intermediate'.

bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("Inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("Active")
  else:
    bioactivity_class.append("Intermediate")
```

```python
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

```python
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

```python
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

```python
data_tuples = list(zip(mol_cid, canonical_smiles, standard_value, bioactivity_class))
df3  = pd.DataFrame(data_tuples, columns = ['molecular_chembl_id', 'canonical_smiles', 'standard_value', 'bioactivity_class'])
df3
```

```python
df3.to_csv('bioactivity_preprocessed_data.csv', index = False)
```
